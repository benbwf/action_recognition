```python
import os
import glob
from pathlib import Path
import collections
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
```

```python
video_dir = 'C:\\Users\\Admin\\dev\\data\\ite_dataset\\demo_videos\\fulltest'

timing_file_paths = glob.glob(os.path.join(video_dir, 'Case2*.csv'))
```

```python
timing_file_paths
```

```python

for filepath in timing_file_paths:
    newvideodirpath = os.path.join(os.path.dirname(filepath), 'split')
    Path(newvideodirpath).mkdir(exist_ok = True)
    with open(filepath) as timingfile:
        section_times = collections.defaultdict(list)
        for i, line in enumerate(timingfile.readlines()):
            if i==0:
                continue
#             print(i, line)
            section, start, end = line.split(',')
#             print(section , start, end)
            start, end = float(start)*60, float(end)*60
            section_times[section].append((start, end))
            
#     print(filepath)
#     print(section_times)
            
            video_path = filepath.replace('.timing.csv', '')
            videofilename = os.path.basename(video_path)
            newvideofilename = videofilename.replace('.MP4', f'_{i:02}_{section}.MP4')
            
            newvideofilepath = os.path.join(newvideodirpath, newvideofilename)
            if start == end:
                continue
            ffmpeg_extract_subclip(video_path, start, end, targetname=newvideofilepath)
            print(videofilename, newvideofilepath)
```
